### 1. Pregunta teórica: ¿Qué diferencia existe entre la covarianza y la correlación?

- La `covarianza` indica si ambas variables varían en la misma dirección (**covarianza positiva**, es decir, si aumenta el valor de una variable aumenta el variable de la otra) o en dirección opuesta (**covarianza negativa** cuando aumenta el valor de una de las variables, disminuye el de la otra). En este caso valor de la covarianza no es importante, solo el signo (si es positivo o negativo)


- La `correlación` varía entre -1 y 1. Dos variables están asociadas o correlacionadas cuando una variable nos da información acerca de la otra. De la misma que la covarianza nos va a indicar la si existe relación entre dos variables. Al contrario que la covarianza, la correlación no nos indica unicamente la dirección (positiva o negativa), también nos indica la "fuerza" de esa relación. De esta forma: 

- Si el valor de la correlación es 0 significa que no hay relación entre las variables. 

- Si el valor es positivo significa que cuando aumenta el valor de una variable aumenta el de la otra. 

- Si el valor es negativoç significa que cuando disminuye el valor de una variable disminuye el de la otra. 



### 2. Calculad la covarianza y la correlación de nuestro dataset. No nos vale solo con que la calculéis, debéis hacer una interpretación de los resultados obtenidos. Calculad solo la correlación de Pearson.

```python
df.cov()
```

Como ya hemos explicado, la covarianza nos indica el sentido de la variación (positiva o negativa) pero no la intensidad. En nuestro caso, vemos que las variables que nos interesan ('hdlngth', 'skullw' y 'totlngth') varían en el mismo sentido ya que el valor de covarianza es positivo. 

```python
df.corr()
```

En el caso de la correlación, volvemos a analizar nuestras variables de interés y podemos observar que todas las correlaciones son positivas, por lo tanto varían en el mismo sentido. Sin embargo, tenemos diferencias en la intensidad de la relación. Podemos observar que tanto el largo de la cabeza como el largo del cráneo tienen una correlación alta con el largo total del cuerpo. Debido a que la correlación es más alta entre el largo de la cabeza y el largo del cuerpo, hemos decidido quedarnos con el largo de la cabeza como nuestra variable predictora. También hemos visto que la correlación entre el largo de la cola y el largo del cuerpo es relativamente alta, por lo tanto consideramos incluirla dentro de nuestro modelo de predicción de largo de cuerpo total para obtener un mejor ajuste. 

### 3. Con los resultados de la correlación del ejercicio anterior, cread un heatmap.

```python
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

sns.heatmap(df.corr(),
cmap = "YlGnBu", 
mask = mask, # incorporamos la máscara en nuestro plot
annot = True)

plt.savefig('../gráficos/heatmap_cor.png');
```

Aquí podemos ver de manera visual lo explicado anteriormente con la intensidad de los colores de nuestro heatmap. Como ya hemos comentado, en azul oscuro podemos observar una correlación alta y positiva entre el largo de la cabeza y el largo total del cuerpo.
